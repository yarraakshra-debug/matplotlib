```python
import matplotlib.pyplot as plt
Months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
Sales_A = [200, 250, 300, 200, 350, 400]
Sales_B = [180, 220, 260, 300, 330, 370]
Ad_Spend = [10, 20, 30, 40, 50, 60]
Sales_scatter = [100, 150, 200, 260, 300, 360]
Categories = ['electronics', 'clothing', 'food', 'furniture']
revenue = [500, 300, 400, 350]
regions = ['north', 'east', 'west', 'south']
region_sales = [30, 25, 20, 25]
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.plot(Months, Sales_A, marker='o', linestyle='dashed', label='Sales_A')
plt.plot(Months, Sales_B, marker='p', linestyle='dotted', label='Sales_B')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 2)
plt.scatter(Ad_Spend, Sales_scatter)
plt.title("Ad Spend vs Sales")
plt.xlabel("Ad Spend")
plt.ylabel("Sales")
plt.grid(True)
plt.subplot(2, 2, 3)
plt.bar(Categories, revenue)
plt.title("Revenue by Category")
plt.xlabel("Categories")
plt.ylabel("Revenue")
plt.subplot(2, 2, 4)
plt.pie(region_sales, labels=regions, autopct='%1.1f%%')
plt.title("Regional Sales Distribution")
plt.tight_layout()
plt.show()
```
